```python
import cv2
import numpy as np
from jaxtyping import UInt8


# observation handling
def prep_obs(state: UInt8[np.ndarray, "... H W 3"]) -> UInt8[np.ndarray, "... 64 64 3"]:
    "Crops out the status bar and resizes 96x96 car racing observations to 64x64"
    cropped = state[..., :84, :, :]

    if cropped.ndim == 4:
        return np.array([cv2.resize(i, (64, 64)) for i in cropped])
    elif cropped.ndim == 3:
        return np.array(cv2.resize(cropped, (64, 64)))
    else:
        raise ValueError(
            f"Expected input to have 3 or 4 input dims, got {cropped.ndim}: shape={state.shape}"
        )
```

```python
state = np.ones(shape=(1, 1, 96, 96, 3))
prep_obs(state).shape
```
