```python
import gymnasium as gym
```

```python
env = gym.make("CarRacing-v3", render_mode="rgb_array")
state, info = env.reset()
```

```python
# 3 networks
# VAE
# MDN_RNN
# Controller (simple one-layer network)

# VAE/MDN_RNN Training:
# 1. Collect 10,000 rollouts from random policy
# 2. Train VAE to model p(z | x) -> p(x | z)
# 3. Train MDN-RNN to model p(z_t1 | z_t, a_t, h_t) # h_t the hidden state
# 4. Train Controller C to maximise expected cumulative reward
```

# Step 1: Collect a Dataset

```python
# An example rollout
import mediapy as media
import numpy as np

state, info = env.reset()

frames = [env.render()]
terminated = False
truncated = False
cum_reward = 0
while not (terminated or truncated):
    action = np.random.uniform(low=env.action_space.low, high=env.action_space.high)
    state, reward, terminated, truncated, _ = env.step(action)
    frames.append(env.render())
    cum_reward += reward

media.show_video(frames, fps=30)
```
